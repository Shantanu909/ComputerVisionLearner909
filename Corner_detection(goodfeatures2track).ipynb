```python
import numpy
import cv2
import matplotlib.pyplot as plt

from google.colab.patches import cv_imshow
```

```python
im = cv2.imread("/content/Starship.jpg")
im = cv2.resize(im,(0,0),fx=0.3,fy=0.3)
im1 = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
cv_imshow(im1)
```

```python
corners = cv2.goodFeaturesToTrack(im1,100,0.01,10)
corners = numpy.int0(corners)
for i in corners:
  x,y = i.ravel()
  cv2.circle(im,(x,y),2,(0,0,255),1)
cv_imshow(im)
```
