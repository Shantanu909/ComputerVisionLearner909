```python
clr = cv2.imread("/content/samp6.png",cv2.IMREAD_COLOR)
cv_imshow(clr)
splitb = clr[::,::,0]
splitg = clr[::,::,1]
splitr = clr[::,::,2]
import numpy as np
modb=np.zeros((splitb.shape[0],splitb.shape[1],3))
modr=np.zeros(((splitb.shape[0],splitb.shape[1],3)))
modg=np.zeros(((splitb.shape[0],splitb.shape[1],3)))
for i in range(splitb.shape[0]):
  for j in range(splitb.shape[1]):
    modb[i,j] = [splitb[i,j],0,0]
    modg[i,j] = [0,splitb[i,j],0]
    modr[i,j] = [0,0,splitr[i,j]]
cv_imshow(splitb)
cv_imshow(splitg)
cv_imshow(splitr)
cv_imshow(modb)
cv_imshow(modg)
cv_imshow(modr)
cv_imshow(modb+modr)
cv_imshow(modb+modg)
cv_imshow(modg+modr)
cv_imshow(modb+modg+modr)
```

```python
b = modb.max()
g = modg.max()
r = modr.max()
for i in range(splitb.shape[0]):
  for j in range(splitb.shape[1]):
    modb[i,j] = b-modb[i,j]
    modg[i,j] = g-modg[i,j]
    modr[i,j] = r-modr[i,j]
splitb = splitb.max()-splitb
splitg = splitg.max()-splitg
splitr = splitr.max()-splitr
cv_imshow(splitb)
cv_imshow(splitg)
cv_imshow(splitr)
cv_imshow(modb)
cv_imshow(modg)
cv_imshow(modr)
```
